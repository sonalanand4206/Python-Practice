1. Calculate Promo Usage Percentage

Add a new column promo_usage_percentage to calculate the percentage of purchases where a promo code was used (Promo Code Used) 
relative to total purchases.

Use apply with lambda to calculate Yes = 100% and No = 0%.

```python
df['promo_usage_percentage'] = df['Promo Code Used'].apply(lambda x : '100' if x == 'Yes' else '0')
df
```

```python
percentage_of_purchases_with_promo = (promo_purcase / total_purchase) * 100
percentage_of_purchases_with_promo
```

2.Categorise High and Low Value Purchases

Create a column purchase_value_category where:
"High Value" if Purchase Amount (USD) > 50.
"Low Value" if Purchase Amount (USD) <= 50.

```python
df['purchase_value_category'] = df['Purchase Amount (USD)'].apply(lambda x : 'High Value' if x > 50 else 'Low Value')
df
```

3.Check Payment Method → Check Subscription Status

Add a column is_subscribed that is True if Subscription Status is "Yes", otherwise False.

```python
df['is_subscribed'] = df['Subscription Status'].apply(lambda x : x == 'Yes')
df
```

5. Calculate Spending Per Purchase

Add a column spending_per_item by dividing Purchase Amount (USD) by Previous Purchases.
Ensure to handle cases where Previous Purchases = 0 to avoid division by zero errors.

```python
(df['Previous Purchases']=='0').count()
```

```python
df['spending_per_item'] = df.apply(lambda x : x['Purchase Amount (USD)'] / x['Previous Purchases'] if x['Previous Purchases'] > 0 else 0, axis = 1)
df
```

```python
df[df['spending_per_item'] < '0']
```

6. Categorise Purchases by Season

Create a column seasonal_category where:
"Winter Lover" if Season = Winter.
"Summer Fan" if Season = Summer.
"Other Seasons" for Spring and Fall.

```python
df['seasonal_category'] = df['Season'].apply(lambda x : 'Winter Lover' if x == 'Winter' else 'Summer Fan' if x == 'Summer' else 'Other Seasons')
df
```

7. Determine Peak Purchase Times

Create a column is_peak_purchase_season where:
True if the Season is "Winter" or "Spring".
False otherwise.

```python
df['is_peak_purchase_season'] = df['Season'].apply(lambda x : (x == 'Winter' or x == 'Spring'))
df
```

8. Flag Unusual Discounts

Add a column unusual_discount where:
True if Discount Applied is "Yes" and Promo Code Used is "No".
False otherwise.

```python
df['unusual_discount'] = df.apply(lambda x : x['Discount Applied'] == 'Yes' and x['Promo Code Used'] == 'No', axis = 1)
df
```

9. Calculate Payment Efficiency

Add a column spending_efficiency that divides Purchase Amount (USD) by Frequency of Purchases.

Categorise it as:
"Low" if spending_efficiency < 10.
"Medium" if 10 <= spending_efficiency < 50.
"High" if spending_efficiency >= 50.

```python
# Map frequency descriptions to numeric values
frequency_mapping = {
    "Weekly": 52,
    "Bi-Weekly": 26,
    "Fortnightly": 26,
    "Monthly": 12,
    "Quarterly": 4,
    "Annually": 1
}
df['frequency_numeric'] = df['Frequency of Purchases'].map(frequency_mapping)

# Calculate spending efficiency
df['spending_efficiency'] = df['Purchase Amount (USD)'] / df['frequency_numeric']

# Categorise spending efficiency
df['spending_efficiency_category'] = df['spending_efficiency'].apply(
    lambda x: 'Low' if x < 10 else 'Medium' if x < 50 else 'High'
)
df
```

```python
df
```

10. Detect Repeated Payment Methods---

Add a column repeated_payment_method where:
True if Payment Method matches Preferred Payment Method.
False otherwise.

```python
df['repeated_payment_method'] = df.apply(lambda x : x['Payment Method'] == x['Preferred Payment Method'], axis = 1)
df
```

```python
df['repeated_payment_method'].value_counts()
```

11. Region and Subscription Analysis---

Create a column region_subscription_analysis where:
    
"Frequent Buyer" if Location is "California" and Frequency of Purchases is "Weekly".
"Occasional Subscriber" if Subscription Status = Yes and Frequency of Purchases is "Annually".
"General" otherwise.

```python
df['region_subscription_analysis'] = df.apply(lambda x : 'Frequent Buyer' if x['Location'] == 'California' and x['Frequency of Purchases'] == 'Weekly' 
                                              else 'Occasional Subscriber' if x['Subscription Status'] == 'Yes' and x['Frequency of Purchases'] == 'Annually' 
                                              else 'General', axis = 1)
df
```

12. Detect High Spending Outliers---

Add a column is_spending_outlier where:
True if Purchase Amount (USD) is more than 3 standard deviations above the mean.
False otherwise.


```python
mean = df['Purchase Amount (USD)'].mean()
mean
```

```python
std = df['Purchase Amount (USD)'].std()
std
```

```python
df['is_spending_outlier'] = df['Purchase Amount (USD)'].apply(lambda x : x > mean + 3 * std)
df
```

13.  Identify Repeat Customers

Add a column is_repeat_customer where:
True if Previous Purchases > 10.
False otherwise.

```python
df['is_repeat_customer'] = df['Previous Purchases'].apply(lambda x : x > 10)
df
```
