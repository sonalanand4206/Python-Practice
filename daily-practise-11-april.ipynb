# üåç Healthcare Spending vs Life Expectancy (1970‚Äì2020)

Welcome to this mini data visualisation project! In this notebook, we explore how healthcare spending has evolved over the last 50 years across five major economies ‚Äî and how it relates to life expectancy.

Using clean visuals and insightful questions, we aim to uncover patterns between **investment in healthcare** and **health outcomes** for countries like Germany, France, Great Britain, Japan, and the USA.

---

## üìä Dataset Overview
The dataset includes the following columns:
- **Year** ‚Äì Ranging from 1970 to 2020  
- **Country** ‚Äì Germany, France, Great Britain, Japan, USA  
- **Spending_USD** ‚Äì Per capita healthcare spending (in USD)  
- **Life_Expectancy** ‚Äì Average life expectancy in years  

---

## üéØ Objectives
Our goal is to use visualisation to answer the following six questions:

---

### ‚úÖ Questions Covered in This Notebook:

#### Q1: Line Plot  
**How has healthcare spending changed over time for each country?**  
*Tool used:* `sns.lineplot()`

#### Q2: Line Plot  
**How has life expectancy evolved over time across countries?**  
*Tool used:* `sns.lineplot()` with hue

#### Q3: Scatter Plot  
**Is there a visible correlation between healthcare spending and life expectancy?**  
*Tool used:* `sns.scatterplot()`

#### Q4: Facet Grid (Scatter + Country Separation)  
**How does spending vs life expectancy differ country by country?**  
*Tool used:* `sns.FacetGrid()` with scatterplot

#### Q5: Bar Plot  
**Which country had the highest increase in life expectancy from 1970 to 2020?**  
*Tool used:* `sns.barplot()`

#### Q6: Heatmap  
**Which countries had the most consistent life expectancy gains by decade?**  
*Tool used:* `sns.heatmap()` after grouping by decade

---

Let‚Äôs dive into the visuals and uncover some global health insights üëá


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('healthexp')
df
```

Q1: How has healthcare spending changed over time for each country?

```python
# Q1: Line Plot


# Step : Plot spending trend over time by country

plt.figure(figsize=(12, 6))


sns.lineplot(data=df, x='Year', y='Spending_USD', hue='Country', marker='o')


plt.title("Healthcare Spending per Capita Over Time (1970‚Äì2020)", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Spending (USD)")


plt.grid(True)


plt.tight_layout()
plt.show()
```

Q2: How has life expectancy evolved over time across countries?

```python
# Q2: Line Plot 


# Line plot of life expectancy over time by country

plt.figure(figsize=(12, 6))


sns.lineplot(data=df, x='Year', y='Life_Expectancy', hue='Country', marker='o')


plt.title("Life Expectancy Trends Over Time (1970‚Äì2020)", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Life Expectancy (Years)")


plt.grid(True)


plt.tight_layout()
plt.show()
```

Q3: Is there a visible correlation between healthcare spending and life expectancy?

```python
# Plot the scatter plot


plt.figure(figsize=(10, 6))


sns.scatterplot(data=df, x='Spending_USD', y='Life_Expectancy', hue='Country', style='Country', s=100)


plt.title("Healthcare Spending vs Life Expectancy (1970 & 2020)", fontsize=14, fontweight='bold')
plt.xlabel("Spending per Capita (USD)")
plt.ylabel("Life Expectancy (Years)")


plt.grid(True)


plt.tight_layout()
plt.show()
```

Q4: How does spending vs life expectancy differ country by country?

```python
# Q4: Facet Grid ‚Äì 


# Create FacetGrid for visualising scatter plot per country


g = sns.FacetGrid(df, col="Country", col_wrap=3, height=4, sharex=False, sharey=False)
g.map_dataframe(sns.scatterplot, x="Spending_USD", y="Life_Expectancy", hue="Year", palette="viridis", s=100)



# Add titles and layout styling
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Spending per Capita (USD)", "Life Expectancy (Years)")


g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Spending vs Life Expectancy ‚Äì Country-wise Comparison", fontsize=14, fontweight='bold')

plt.show()
```

Q5: Which country had the highest increase in life expectancy from 1970 to 2020?

```python
# Q5: Bar Plot ‚Äì 


# Step 1: Filter data for 1970 and 2020 only
df_filtered = df[df['Year'].isin([1970, 2020])]


# Step 2: Pivot the data to calculate difference between 1970 and 2020
life_exp_diff = df_filtered.pivot(index='Country', columns='Year', values='Life_Expectancy').reset_index()
life_exp_diff['Increase'] = life_exp_diff[2020] - life_exp_diff[1970]


# Step 3: Sort and plot
plt.figure(figsize=(10, 6))
sns.barplot(data=life_exp_diff.sort_values(by='Increase', ascending=False), x='Country', y='Increase', palette='crest')
plt.title("Increase in Life Expectancy from 1970 to 2020", fontsize=14, fontweight='bold')
plt.ylabel("Years Gained")
plt.xlabel("Country")
plt.tight_layout()
plt.show()


# Display the table for reference
life_exp_diff.sort_values(by='Increase', ascending=False)
```

Q6: Which countries had the most consistent life expectancy gains by decade?

```python
# Q6: Heatmap ‚Äì Which countries had the most consistent life expectancy gains by decade?

# Step 1: Create a Decade column
df['Decade'] = (df['Year'] // 10) * 10

# Step 2: Calculate average life expectancy by country and decade
heatmap_data = df.groupby(['Country', 'Decade'])['Life_Expectancy'].mean().reset_index()

# Step 3: Pivot the data to make countries rows and decades columns
heatmap_matrix = heatmap_data.pivot(index='Country', columns='Decade', values='Life_Expectancy')

# Step 4: Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_matrix, annot=True, cmap='YlGnBu', fmt=".1f", linewidths=0.5, linecolor='gray')
plt.title("Average Life Expectancy by Country and Decade", fontsize=14, fontweight='bold')
plt.xlabel("Decade")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
