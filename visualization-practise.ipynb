Graphs :-

1. Line chart   : plt.plot
2. Bar chart    : plt.bar
3. Pie chart    : plt.pie
4. Box Plot     : df.boxplot
5. Histogram    : df.hist
6. Scatter Plot : plt.scatter

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
d={"IPL_Year":[2008,2009,2010,2011,2012,2013,2014,2015,2016,2017], "Rohit":[404,362,404,372,433,538,390,482,489,333], "Virat":[165,246,307,557,364,639,359,505,973,308]}
df=pd.DataFrame(d)
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize = (8,4))      # mandatory to write.
plt.plot(df['IPL_Year'],df['Rohit'],marker = 'D',label='Rohit Sharma')
plt.plot(df['IPL_Year'],df['Virat'],marker = 'D',label = 'Virat Kohli')
plt.legend(fontsize = 12,loc = 'upper center')
plt.show()     # mandatory to write.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize = (8,4))         # mandatory to write.
plt.plot(df['IPL_Year'],df['Rohit'],marker = 'D',label='Rohit Sharma',color = 'Orange')
plt.plot(df['IPL_Year'],df['Virat'],marker = 'D',label = 'Virat Kohli',color = 'yellow')
plt.legend(fontsize = 12,loc = 'upper center')
plt.title('IPL Performance',fontweight = 'bold',fontsize = '18',color = 'Brown',y=1.2)
plt.show()        # mandatory to write.
```

```python
# plt.grid

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize = (8,4),dpi = 120)               #DPI - dots per inch

plt.plot(df['IPL_Year'],df['Rohit'],marker = 'D',label='Rohit Sharma',color = 'Orange')
plt.plot(df['IPL_Year'],df['Virat'],marker = 'D',label = 'Virat Kohli',color = 'yellow')
plt.legend(fontsize = 12,loc = 'upper center')
plt.title('IPL Performance',fontweight = 'bold',fontsize = '18',color = 'Brown',y = 1.1)
plt.xlabel('Years',color = 'Blue')
plt.ylabel('Runs',color = 'Red')
plt.grid(True, color = 'black',linestyle = '--')

plt.show()





#.svg , .pdf , .png , .jpeg
```
