```python
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

```python
# o Analyze the total quantity of each mineral produced per year across all districts.


pd.pivot_table(df,index = 'Mineral', columns = 'Year', values = 'Quantity (Tonnes)', aggfunc = 'sum')
```

```python
# o Identify which mineral had the highest production in each year.


pd.pivot_table(df, index = 'Mineral', columns = 'Year', values = 'Quantity (Tonnes)', aggfunc = 'max').idxmax()
```

Explanation:



pd.pivot_table(...): Creates a pivot table summing the quantities of minerals for each year.


                                                                             
.idxmax(): Directly identifies the index (mineral) corresponding to the maximum value in each year.

    8.	Revenue-to-Quantity Efficiency


o	Calculate the revenue-to-quantity ratio for each company and mineral combination.

    
o	Identify the most efficient companies in terms of revenue generated per tonne of mineral.

```python
#  o Calculate the revenue-to-quantity ratio for each company and mineral combination.


result = pd.pivot_table(df,index = 'Mineral', columns = 'Company', values = ['Revenue (Million INR)','Quantity (Tonnes)'], aggfunc = 'sum')
result.head(2)
```

```python
ratio = result['Revenue (Million INR)'] / result['Quantity (Tonnes)']
ratio
```

```python
ratio.fillna(0, inplace=True)
ratio
```

```python
# 2nd attempt using groupby().

#  o Calculate the revenue-to-quantity ratio for each company and mineral combination.


grouped = df.groupby(['Company','Mineral'])[['Revenue (Million INR)','Quantity (Tonnes)']].sum()
grouped
```

```python
# different approach using normal new row creation and calculating the ratio.

grouped['Revenue to Quantity Ratio'] = grouped['Revenue (Million INR)'] / grouped['Quantity (Tonnes)']
grouped.head(2)
```

```python
# different approach using lambda with apply and creating a new column.


grouped['Revenue to Quantity Ratio'] = grouped.apply(lambda x : x['Revenue (Million INR)'] / x['Quantity (Tonnes)'], axis = 1)
grouped
```

    9.	Mineral Production Dominance Over Time


o	Determine which mineral was the most produced for each year.

                                                 
o	Identify shifts in mineral dominance over the years.

```python
#  o Determine which mineral was the most produced for each year.


most_produced = pd.pivot_table(df,index = 'Mineral', columns = 'Year', values = 'Quantity (Tonnes)', aggfunc = 'sum').idxmax()
most_produced
```

```python
#  o Identify shifts in mineral dominance over the years.


shifts = most_produced[most_produced != most_produced.shift()]
shifts
```
